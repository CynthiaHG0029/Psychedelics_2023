```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = "https://psychedelicalpha.com/data/psychedelic-drug-development-tracker"
response = requests.get(url)
```

En este código busco hacer un df de los textos del pop up de cada lab

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
div_elements = soup.find_all('div', class_='uael-content')
```

```python

text_list = []
for div_element in div_elements:
    p_element = div_element.find('p')
    text = p_element.get_text(strip=True)
    text_list.append(text)
```

```python
text_list
```

```python
div_df = pd.DataFrame({'Text': text_list})
```

```python
div_df
```

```python
url = "https://psychedelicalpha.com/data/psychedelic-drug-development-tracker"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")

alt_text_list = []
images = soup.find_all("img")
for image in images:
    alt_text = image.get("alt")
    alt_text_list.append(alt_text)

df_d = pd.DataFrame(alt_text_list, columns=["Alt Text"])
print(df_d.head())
```

```python
labs = df_d.iloc[:89,:]
```

```python
labs
```

```python
labs['Alt Text'] = labs['Alt Text'].str.replace(r'logo?.\w*', '', regex=True)
```

```python
labs
```

```python
labs['Alt Text'] = labs['Alt Text'].str.replace(r'Logo?.\w*', '', regex=True)
```

```python
#labs.drop(index=labs.index[0], axis=0, inplace=True)
```

```python
labs
```

```python
labs['Alt Text'] = labs['Alt Text'].shift(-1)
```

```python
labs
```

```python
labs_div=pd.merge( labs, div_df, left_index=True, right_index=True)
```
