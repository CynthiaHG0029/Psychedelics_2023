```python
import os
import re
import pandas as pd


import nltk
nltk.download("punkt")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer

import string

from collections import Counter

```

```python
data = pd.read_csv('./Psilocybin_march08.csv')
data.columns.to_list()
```

```python
data_p = data.rename(columns={'General Subject Matter':'general_subject'})
```

```python
stops = set(stopwords.words("english"))
black_list = ["methods", "disorder", "major","use", "compositions", "method", "treatment", "disorders", "derivative",
               "comprising","thereof", "treating", "formulations", "compounds",
               "preparation","receptor", "combination", "using","systems", "uses", "production",
               "agents", "disorder","intermediates","composition","forms","delivery", "enhancing", "pharmaceutical"  ]
for o in black_list:
  stops.add(o)
print(stops)
```

```python
def tokenizer(frase):
  final = []
  sopa = word_tokenize(frase)
  stemmer = SnowballStemmer("english")
  punct = string.punctuation+"¿?¡"
  for palabra in sopa:
    lowerPalabra = palabra.lower()
    if lowerPalabra not in stops and lowerPalabra not in punct:
      lowerPalabra = stemmer.stem(lowerPalabra)
      final.append(lowerPalabra)

  return final
```

```python
import gensim
```

```python
tokens_solos = data_p['tokenized_text'].tolist()
dictionary = gensim.corpora.Dictionary(tokens_solos)
```

```python
dictionary
```

```python
dictionary.filter_extremes(no_below=4, no_above=0.8, keep_n=100000)
```

```python
doc_term_matrix = [dictionary.doc2bow(rev) for rev in tokens_solos]
```

```python
LDA = gensim.models.ldamodel.LdaModel
lda_model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=5, random_state=100, chunksize=1000, passes=50, iterations=100)
```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
pyLDAvis.enable_notebook()
```

```python
vis = pyLDAvis.gensim_models.prepare(lda_model, doc_term_matrix, dictionary)
vis
```
